```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_tracks = pd.read_csv("spotify_tracks.csv")
spotify_tracks.head()
```

```python
# data cleaning - get info about where null values are
pd.isnull(spotify_tracks).sum()
```

```python
# show info of the dataframe
spotify_tracks.info()
```

```python
# remember sort_values("COLUMN"), not sort.values
sorted_tracks_popularity = spotify_tracks.sort_values("popularity",
                          ascending = True).head(10)
sorted_tracks_popularity
```

```python
# summary statistics
spotify_tracks.describe()
```

```python
# query function queries (filters) values by boolean expressions
most_popular_1 = spotify_tracks.query("popularity > 90",inplace = False)
most_popular_1 = most_popular_1.sort_values("popularity",ascending = False)

# if you want to specify more than one column, use double square brackets
most_popular_bar_data = most_popular_1[:10][["popularity","artists","name"]]
most_popular_bar_data
```

```python
sns.barplot(data = most_popular_bar_data,
            x = "artists",
            y = "popularity")
plt.xticks(rotation = 45,
          fontsize = 4)
```

```python
# query function queries (filters) values by boolean expressions
most_popular = spotify_tracks.query("popularity > 90",inplace = False)
most_popular = most_popular.sort_values("popularity",ascending = False)
```

```python
# making an existing column an index *index is not column!
# with inplace = True, the dataframe will make the change permanent
spotify_tracks.set_index("release_date",inplace = True)
spotify_tracks.index = pd.to_datetime(spotify_tracks.index)
```

```python
# you can also reset index using reset_index function
spotify_tracks.reset_index()
```

```python
# Since duration is milli seconds, convert them to seconds
# apply function applies a function along "an axis" of the DataFrame.
spotify_tracks["duration"] = spotify_tracks["duration_ms"].apply(lambda x: round(x/1000))
spotify_tracks.drop("duration_ms",inplace = True,axis = 1)
spotify_tracks
```

```python
# check the if duration_ms was dropped and duration was created
spotify_tracks.duration.head()
```

```python
# Creating another column of the duration_min
spotify_tracks["duration_min"] = spotify_tracks["duration"].apply(lambda b: round(number = b/60,ndigits = 1))
spotify_tracks["duration_min"]
```

```python
# what is the longest song?
spotify_tracks.sort_values(["duration"],ascending = False)
```

```python
# dropping unnecessary variables
correlation_tracks = spotify_tracks.drop(["key","mode","explicit","duration"],axis = 1)
```

```python
# using Peason Correlation Coefficient
correlation_tracks = correlation_tracks.corr(method = "pearson")
```

```python
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(correlation_tracks,
                      annot = True,
                      fmt = ".1g",
                      vmin = -1,
                      vmax = 1,
                     center = 0,
                     cmap = "flare",
                     linewidths = 1,
                     linecolor = "Black")
heatmap.set_title("Data: Spotify")
heatmap.set_xticklabels(heatmap.get_xticklabels(),
                       rotation = 45)
plt.suptitle("Correlation Heatmap among Variables",size = 20)
plt.savefig("Heatmap.png")
```

```python
# samping the dataset
# sample(int(percentage*len(spotify_tracks)))
# this case, the function is sampling 0.4% of the dataset

spotify_sample = spotify_tracks.sample(int(0.004*len(spotify_tracks)))
```

```python
print(len(spotify_sample))
```

```python
# setting the size of the plot (dimentions)
plt.figure(figsize = (8,6))
# sns.regplot plots regular plot on Python.

sns.regplot(data = spotify_sample,
           x = "acousticness",
           y = "energy",
           color = "lightblue").set(title = "Data: Spotify")
plt.suptitle("Correlation between Loudness and Energy",
            size = 15)
```

```python
# making another column that is only year data!
# First, (pandas!) index.get_level_values function get an individual level of values from a multiindex
# telling Python to get "release date" index and make a new column named "date" 
spotify_tracks["date"] = spotify_tracks.index.get_level_values("release_date")
```

```python
# Next, making another dataset to convert date column to datetime class
# pandas.to_datetime function is useful here
spotify_tracks.date = pd.to_datetime(spotify_tracks.date)
```

```python
# Lastly, make another dataset named years using pandas.dt.year function
# this extracts only year part of the datetime data!
years = spotify_tracks.date.dt.year
```

```python
#pip install --user seaborn==0.11.0
```

```python
# making a distribution plot
sns.displot(years,
            discrete = True,
            aspect = 2,
            height = 5,
            kind = "hist",
           color = "green")
plt.title("the Number of Songs Made Per Year",size = 20)
plt.suptitle("Data: Spotify.Inc",
            x = 0.5,
            y = 0.94)
plt.xlabel("Year")
```

```python
total_duration = spotify_tracks.duration_min
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years,
                  y = total_duration,
                  ax = ax,
                  errwidth = False)
plt.title("Total Duration of the Songs based on Year",size = 20)
plt.xticks(rotation = 90)
plt.xlabel("Year")
plt.ylabel("Song Duration in Minutes")
plt.show()
```
